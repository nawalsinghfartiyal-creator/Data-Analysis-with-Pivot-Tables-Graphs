## The re module

```python
import re
```

### re.search()

```python
text = """Hellow World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
```

```python
# Return the first match
re.search(r'\d' , text)
```

```python
# Flags
re.search(r'^hellow' , text, flags=re.I)
```

### re.findall()

```python
# Find all the matches
re.findall(r'\d' , text)
```

#### Exercise

```python
# Match only puncuation in text 
re.findall(r'[^\w\s]' , text)
```

```python
# Match the right date format (MM-DD-YYYY)
text= """
13-04-2021
2021-13-04
2021-04-13
"""
```

```python
re.findall(r'\d{2}-\d{2}-\d{4}' , text)
```
